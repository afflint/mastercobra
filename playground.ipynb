## Funzioni

```python
def sm(s=[]):
    """
    This function calculates sum and mean
    @param: s is a list that...
    """
    a = 0
    for x in s:
        a += x
    try:
        m = a / len(s)
    except ZeroDivisionError:
        m = None
    return a, m
```
