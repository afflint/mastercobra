```python
def thousands(x):
    if pd.isnull(x):
        return x
    else:
        try:
            return float(x.replace(',', ''))
        except ValueError:
            return np.nan
        
def perc(x):
    if pd.isnull(x):
        return x
    else:
        try:
            return float(x.replace('%', '')) / 100
        except ValueError:
            return np.nan
    
```

```python
conversion_map = {
    3: thousands, 6: thousands, 7: thousands, 9: thousands, 10: thousands, 11: thousands,
    12: thousands, 13: thousands, 5: perc
}
```

```python
df2 = pd.read_csv(f, converters=conversion_map)
```

```python
df2.head()
```

```python
df2.dtypes
```
